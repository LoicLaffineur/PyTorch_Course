### Reshaping, view, stack, (un)squeeze, permute :  

```python
## reshaping, stacking and squeeze/unsqueeze

## view : return a view of an imput tensor of certain shape but keep the same memory

## stacking : combine multiple tensors on top of each other or side by side

## squeeze : removes all "1" dim from tensor

## unsqueeze : add a "1" dim to a target tensor

## permute : return a view of the input with dimensions permuted in a certian way
```

```python
x= torch.arange(1.,10.)
x, x.shape
```

```python
# reshape
x.reshaped=x.reshape(1,9)
x.reshaped, x.reshaped.shape
```

```python
# view :
z=x.view(1,9)
z, z.shape
```

```python
# changing z changes x bc view shares the same memory
z[:,0]=5
x,z
```

```python
# stack

x_stacked=torch.stack([x,x,x,x],dim=0)
x_stacked,x_stacked.shape
```

```python
# squeeze and unsqueeze

x.reshaped.shape, x.reshaped.squeeze(), x.reshaped.squeeze().shape

x_squeeze=x.reshaped.squeeze()

x_squeeze.shape, x_squeeze.unsqueeze(dim=1), x_squeeze.unsqueeze(dim=1).shape
```

```python
# permute :
x_orig=torch.rand(size=(224,224,3))

x_permuted=x_orig.permute(2,0,1) # permute est une view !!!
x_orig.shape, x_permuted.shape
```

## PyTorch tensor & numpy :

```python
## numpy array to tensor
import numpy as np

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array)
array, tensor # default type of array is float64
```

```python
array=array+1
array, tensor # doesn't change
```

```python
# tensor to numpy array

tensor= torch.ones(7)
numpy_tensor=tensor.numpy()

tensor, numpy_tensor # keep the datatype of float32 from torch !!
```

```python
tensor=tensor+1
tensor,numpy_tensor # doesn't change
```
